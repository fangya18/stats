### Chi-square test application 1:
#### Test Goodness of the Fit

test if the observed categorical data follows the hypothesizedÂ or expected distribution.

from scipy.stats import chi2
from scipy.stats import chisquare
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns


#### Example 1, p-value
Suppose we f_exp are the expect number of boys we would like to see in
5 classrooms. f_obs are the observed number of boys in 5 classrooms. We want to see if it is the same as the distribution.
H0(Null) : the observation distrubtion is consistent with the expected distrubtion


```python
chisquare(f_obs=[18,15,5,8,4,3], f_exp=[10,5,7,18,10,11])
```

As the p-value showed. It is signifacantly smaller than 0.05, so we reject
the N0, and conclude the f_obs is a different distribution from f_exp.

#### Example 2, Plot

We will graph a Chi-square distribution with degree 5 and size 1000, and use Kernel Density Estimation to fit the graph.

```python
r=chi2.rvs(5, size=1000)

#histogram
plt.hist(r, normed=False , alpha=0.5)
plt.title("Chi-square distribution with degree 5")
plt.xlabel("Value")
plt.ylabel("Frequency Count")
plt.show()

#estimate of distribution using Kernel Density Estimation
sns.kdeplot(r)

#Combine Histogram and KDE
sns.distplot(r)
plt.title("Chi-square Dist(df=5)with KDE")
plt.xlabel("Values")
plt.ylabel("Density")

plt.show()
```
